```python

!pip install prophet plotly matplotlib pandas scikit-learn --quiet


import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import plotly.graph_objects as go
from google.colab import files


print("ðŸ“¤ Upload your sales CSV file...")
uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])


df['date'] = pd.to_datetime(df['date'])
df_prophet = df.rename(columns={'date': 'ds', 'sales': 'y'})


model = Prophet(seasonality_mode='multiplicative')
model.fit(df_prophet)


future = model.make_future_dataframe(periods=90)
forecast = model.predict(future)


fig1 = model.plot(forecast)
plt.title("Sales Forecast")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()


fig2 = model.plot_components(forecast)
plt.show()

split = int(len(df_prophet) * 0.8)
train = df_prophet[:split]
test = df_prophet[split:]

m = Prophet()
m.fit(train)
future_test = m.make_future_dataframe(periods=len(test))
forecast_test = m.predict(future_test)
pred = forecast_test['yhat'].iloc[-len(test):].values
true = test['y'].values

mae = mean_absolute_error(true, pred)
rmse = np.sqrt(mean_squared_error(true, pred))

print("ðŸ“Š Model Evaluation")
print("MAE:", mae)
print("RMSE:", rmse)

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=df_prophet['ds'], y=df_prophet['y'],
    mode='lines', name='Actual Sales'
))
fig.add_trace(go.Scatter(
    x=forecast['ds'], y=forecast['yhat'],
    mode='lines', name='Forecast'
))
fig.update_layout(
    title="Interactive Sales Forecast Dashboard",
    xaxis_title="Date",
    yaxis_title="Sales"
)
fig.show()

print("\nâœ… Dashboard generation completed successfully!")
```
